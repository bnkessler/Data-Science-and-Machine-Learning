#### Question 2.1 [30 points]

Write a functions to help recommend trails to a user. The function will have as its input the dataset and the name of a National Park, and the following optional arguments:

    A minimum trail elevation gain (in meters)
    A maximum trail elevation gain (in meters)
    A minimum trail rating
    The type of trail


```python
data.head()
```

```python
def recommend_trails(dataset, national_park, min_elev = None, max_elev = None, min_rating = None, trail_type = None):
  
    if min_elev is None:
        min_elev = min(dataset["elevation_gain"])
    if max_elev is None:
        max_elev = max(dataset["elevation_gain"])
    if min_rating is None:
        min_rating = min(dataset["avg_rating"])
    if trail_type is None:
        trail_type = ["out and back", "loop", "point to point"]
    
    recs = pd.DataFrame(columns=dataset.columns)
    recs2 = pd.DataFrame(columns=dataset.columns)
    recs3 = pd.DataFrame(columns=dataset.columns)
    rec = dataset[dataset["area_name"] == national_park]
    for elev in rec["elevation_gain"]:
        if min_elev <= elev <= max_elev:
            recs = pd.concat([recs, rec.loc[rec["elevation_gain"] == elev]])
    for rating in recs["avg_rating"]:
        if rating >= min_rating:
            recs2 = pd.concat([recs2, recs.loc[recs["avg_rating"] == rating]])
    for t_type in recs2["route_type"]:
        if t_type in trail_type:
            recs3 = pd.concat([recs3, recs2.loc[recs2["route_type"] == t_type]])

    return recs2
```

```python
recommend_trails(data, "Haleakala National Park", min_elev = 1000, min_rating = 4)
```

#### Question 2.2 [30 points]

It’s reasonable to think we might also want to search for hikes according to climate. Therefore, we will scrape the average temperatures for each U.S. state from the following website:

    https://www.extremeweatherwatch.com/us-state-averages

Write a function that takes one optional arguent, month.

The function should scrape the temperature website above for the desired month. If no month is specified then it should scrape the overall average temperatures. The function should return a data frame of the temperatures for the specified month in the names of the temperature columns.

Here is some starter code for this function:

```python
import requests
from bs4 import BeautifulSoup
def scrape_temps(month = None):

    my_url = "https://www.extremeweatherwatch.com/us-state-averages"
    if month == None:
        response = requests.get(my_url)
    else:
        my_url = "https://www.extremeweatherwatch.com/us-state-averages/month-" + month
        response = requests.get(my_url)
    
    soup = BeautifulSoup(response.content, "html.parser")
    table = soup.find_all("td")
    data = []
    state = []
    tempf = []
    tempc = []
    df = pd.DataFrame(columns = ["State", "Temp F", "Temp C"])
    for row in table:
        data.append(row.get_text().split(" "))

    df["State"] = data[0::4]
    df["Temp F"] = data[1::4]
    df["Temp C"] = data[2::4]
        

    return df
```

```python
scrape_temps(month = "january")
```
